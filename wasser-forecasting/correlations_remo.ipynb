```python
import pandas as pd
import json
import matplotlib.pyplot as plt



with open('data/266322010.json', 'r') as file:
    json_data = json.load(file)
parsed_data = json_data[0]["data"]  # Extracting the "data" list
df = pd.DataFrame(parsed_data, columns=["Timestamp", "Value"])

df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
df
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Timestamp'], df['Value'], label='Value', color='blue')


# Format the plot
plt.title('Time Series Plot')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.grid(True)
plt.legend()

# Format the plot
plt.title('Time Series Plot')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```
